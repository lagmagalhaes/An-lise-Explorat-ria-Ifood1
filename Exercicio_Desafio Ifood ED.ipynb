O Salário mais alto é de 113.734


```python
# criamos uma coluna que define quem ganha bem, e quem não ganha bem, tendo como base o 3º quartil: 68281.0
df['Categoria_Salario'] = np.where(df['Income']>68281.0, "ganham_bem", "nao_ganham_bem")
df
```

```python
#Contagem de quem 'ganha_bem' e de quem 'não_ganha_bem'
df['Categoria_Salario'].value_counts()
```

```python
# Contagem dos valores de 'Categoria_Salario'
categoria_salario_counts = df['Categoria_Salario'].value_counts()

# Criação do gráfico de barras com quem 'ganha_bem' e de quem 'não_ganha_bem'
sns.barplot(x=categoria_salario_counts.index, y=categoria_salario_counts.values)
plt.xlabel('Categoria do Salário')
plt.ylabel('Valores')
plt.xticks(rotation=90)
plt.show()
```

```python
 # Histograma do salário do cliente
df["Income"].hist()
```

```python
import plotly.express as px
fig = px.histogram(df, x="Income", nbins=15, title="Distribuição do Salário")
fig.show()
```

```python
# Boxplot Income
sns.boxplot(df["Income"])
```

```python
# Boxplot Distribuição do Salário
import plotly.express as px
fig = px.box(df, y="Income", title="Distribuição do Salário")
fig.show()
```

Com o boxplot Income e o Boxplot Distribuição do Salário verificamos que não existem outliers

```python
# Calculo do coeficiente de variação
df["Income"].std()/df["Income"].mean()
```

```python
df["Income"].skew()
```

 Pelo valores de skew e pelo boxplot, podemos dizer que apesar de não completamente normal, a curva se aproxima de uma normal.

# Que nivel de educação tem os nossos clientes

```python
# Gráfico de Barras com os niveis de educação
sns.countplot(x=df["education_level"])
```

```python
#Gráfico de barras paralelas para uma melhor visualização
plt.figure(figsize=(10, 6))
sns.countplot(x='education_level', hue='Categoria_Salario', data=df)

plt.title('Relação entre Nível de Educação e Categoria de Salário')
plt.xlabel('Nível de Educação')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=90)
plt.legend(title='Categoria de Salário')

plt.show()
```

# Qual o estado civil dos nossos clientes

```python
estado_civil = df['marital_status'].value_counts()
print(estado_civil)

sns.barplot(x=estado_civil.index, y=estado_civil.values)
plt.xlabel('Estado civil')
plt.ylabel('Valor')
plt.title('Valor de níveis de educação')
plt.xticks(rotation=90)
plt.show()
```

```python
# Quantos clientes temos em cada estado civil?
# Gráfico de Barras com o estado civil
sns.countplot(x=df["marital_status"])
```

```python
# Qual é a relação do estado civil e o número de filhos?
# Será que as pessoas casadas têm mais filhos?
sns.boxplot(x='marital_status', y='kids', data=df)
```

```python
df.groupby(["marital_status"])["kids"].describe()
```

```python
# Calculo da Média de filhos por estado civil
df.groupby("marital_status")["kids"].mean()
```

```python
#Calculo da Mediana de filhos por estado civil
df.groupby("marital_status")["kids"].median()
```

```python
# Qual a relação entre as pessoas gastarem mais ou menos na plataforma quando se têm filhos?
sns.boxplot(x='kids', y='expenses', data=df)
```

```python
#Calculo da Média de Gastos
df.groupby("kids")["expenses"].mean()
```

```python
# Calculo da Mediana
df.groupby("kids")["expenses"].median()
```
